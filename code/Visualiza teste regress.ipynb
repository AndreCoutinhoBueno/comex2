```python
##### Computação cietífica
import numpy as np
##### Manipulação e análise
import pandas as pd
##### Análise exploratória visual
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
%store -r tpr_g_t
tpr_g_t.grupo_mes.unique()
```

```python
tpr_g_t=tpr_g_t[['date','NCM','grupo_mes','KG','KG_pred']].sample(frac=.2)
```

```python
tpr_g_t.date=tpr_g_t.date.astype('datetime64[ns]')
```

```python
tpr_g_t_l=pd.concat([tpr_g_t.loc[:,:'KG'].assign(tipo='real'),tpr_g_t[['date','NCM','grupo_mes','KG_pred']].rename(columns={'KG_pred':'KG'}).assign(tipo='pred')])
```

```python
tpr_g_1=tpr_g_t_l[tpr_g_t_l.grupo_mes==1]
```

```python
tpr_g_2=tpr_g_t_l[tpr_g_t_l.grupo_mes==2]
```

```python
tpr_g_3=tpr_g_t_l[tpr_g_t_l.grupo_mes==3]
```

```python
sns.relplot(data=tpr_g_1,x='date',y='KG', hue='tipo',col='NCM',col_wrap=3).fig.suptitle('Testes por Mês',y=1.05)
plt.show()
```

```python
sns.relplot(data=tpr_g_2,x='date',y='KG', hue='tipo',col='NCM',col_wrap=3).fig.suptitle('Testes por Bimestre',y=1.05)
plt.show()
```

```python
sns.relplot(data=tpr_g_3,x='date',y='KG', hue='tipo',col='NCM',col_wrap=3).fig.suptitle('Testes por Trimestre',y=1.05)
plt.show()
```
